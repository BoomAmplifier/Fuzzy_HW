#### Problem 1 (Ch5-15):  
Show that the following two nonlinear models are intrinsically linear:    
- a)$y=\frac{1}{1+e^{\frac{ax}{b+x}}}$      
- b)$y=\ln a +x- \ln(b+e^x)$     
(In both cases, $a$ and $b$ are fitting parameters.)    

**Solution:**   
- a) $y=\frac{1}{1+e^{\frac{ax}{b+x}}}$

$$\begin{equation}
\begin{array}{rcl}
\frac{1}{y} &=& 1+e^{\frac{ax}{b+x}} \\
\frac{1}{y} -1 &=& e^{\frac{ax}{b+x}} \\
\ln {\frac{1}{y} -1} &=& \frac{ax}{b+x} \\
\frac{1}{\ln {y^{-1} -1}} &=& \frac{b+x}{ax} \\
                         &=& \frac{b}{a}x^{-1} + \frac{1}{a}
\end{array}
\end{equation}$$

- b) $y=\ln a +x- \ln(b+e^x)$

$$\begin{equation}
\begin{array}{rcl}
e^y &=& e^{\ln a +x- \ln(b+e^x)} \\
e^y &=& e^{\ln a}  \cdot e^{- \ln(b+e^x)} \cdot e^{x} \\
e^y &=& a\cdot (b+e^x)^{-1} \cdot e^{x} \\
e^y &=& \frac{ae^{x}}{b+e^x} \\
e^{-y} &=& \frac{b+e^x}{ae^{x}} \\
       &=& \frac{b}{a}e^{-x} + \frac{1}{a}
\end{array}
\end{equation}$$

Let new $\tilde{y} = e^{-y}, \tilde{x} = e^{x}$, the system model can be written as
$$ \tilde{y} = \frac{b}{a}\tilde{x}^{-1} + \frac{1}{a}$$
Which means the nonlinear model is intrinsically linear.
