- 2.Applying method proposed in [[1]](#r1) (**a beautiful way**)
    
> According to [[1]](#r1), we divide $x\in[0,6.5]$ into 100 equal intervals, the piecewise linear function $S$ may defined as $\text{supp}(S)\subset [0,6.5]$ and $||S-f|| < \xi/2$. Let $A_j(j=0,...,100)$ be fuzzy numbers as follows  
$$\begin{equation}
A_0(t)=
\left\{
\begin{array}{}
50(\frac{1}{50}-t) & & 0\leq t \leq \frac{1}{50} \\
0  & & \text{otherwise}
\end{array}
\right.
\end{equation}$$
$$\begin{equation}
A_{100}(t)=
\left\{
\begin{array}{}
50(t-\frac{99}{50}) & & \frac{99}{50}\leq t \leq 6.5 \\
0  & & \text{otherwise}
\end{array}
\right.
\end{equation}$$    
with $j=1,2,...,99$     
$$\begin{equation}
A_j(t)=
\left\{
\begin{array}{}
50(t-\frac{j-1}{50}) & & \frac{j-1}{50} \leq t \leq \frac{j}{50} \\
50(\frac{j+1}{50}-t) & & \frac{j}{50} \leq t \leq \frac{j+1}{50} \\
0  & & \text{otherwise}
\end{array}
\right.
\end{equation}$$   
The Mamdani fuzzy system F as follows
$$ F(t) = \sum_{j=0}^{100}A_j(t)f(\frac{j}{50}) $$

**Discussion:**  
For a SISO fuzzy system, the approximation theory presented in [[1]](#r1) could be considered as a kind of interpolating theory for a continuous function. A rule base of a Mamdani fuzzy system can be described as **If $x$ is $\bar{A}_i$, then $y$ is $\bar{B}_i$.**      
If $\bar{A}_i$ has been choosed as [[1]](#r1) mentioned, $\bar{B}_i$ could be calculated by using a sampling alike theory. Then the continuous function can be apporoached by using $i$ MFs which is $\bar{A}_i$ and $\bar{B}_i$.

### Sugeno fuzzy model   
The same method as we discussed above, a Sugeno fuzzy model for the system can be represent as   
$$\text{If  } x \text{ is in } \bar{C}_i, \text{then  } y =\bar{G}(x).$$
$x\in [0,6.5]$ can be divided into 100 equal sub-intervals using 99 points $x_j$, $j=0,2,...,100$. The input MFs can be chosed as
$$x_i = triangle(x, x_{i-1},\frac{x_{i-1}+x_{i}}{2},x_{i}) \text{ when }x_i \in [x_{i-1},x_{i}] \text{  }(i=1,2,...100)$$,
while output MFs can be chosed as
$$y_i =G_{i} = f(\frac{x_{i-1}+x_{i}}{2})\text{ }(i=1,2,...100)$$
Thus, the fuzzy output should be the same as $f(x)$

**Discussion:** A approximation theory for using Takagi Sugeno fuzzy system to approch Mamdani fuzzy system is discussed in [[2]](#r2). 
> The Mamdani system can be approximated by the TS fuzzy system
$$TS(x)=\frac{\sum_{i=1}^{n}A_i(x)(a_ix+b_i}{sum_{i=1}^{n}A_i(x)}$$
>
> $$a_i=\frac{y_{i+1}-y_{i-1}}{x_{i+1}-x_{i-1}}$$
>
> $$b_i=y_{i}-a_ix_i$$

### Taylor's expansion    
$$f(x)=f(x^0)+\sum_{i=1}^{n}(x_i-x_i^0)\frac{\partial f}{\partial x_i}(x_i^0)+\bar{f}_{x^0}(x)$$
