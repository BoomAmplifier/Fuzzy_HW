## Solutions:  

The fuzzy system described in the project can be represented as:
- 1.Input
$$\begin{equation}
\left\{
\begin{array}{}
Error &= \{ (x,\mu_A(x))| x\in[-1,1]\} \\
Rate &= \{ (y,\mu_B(y))| y\in[-2,2]\}
\end{array}
\right.
\end{equation}$$

$$\begin{equation}
\mu_A(x)=
\left\{
\begin{array}{}
\text{Positive:  } & \text{triangle}(x;0,1,2) \\
\text{Zero:  } & \text{triangle}(x;-1,0,1) \\
\text{Negative:  } & \text{triangle}(x;-2,1,0)
\end{array}  
\right.
\end{equation}$$

$$\begin{equation}
\mu_B(y)=
\left\{
\begin{array}{}
\text{Positive:  } & \text{gaussian}(y;0.8,2) \\
\text{Zero:  } & \text{gaussian}(y;0.8,0) \\
\text{Negative:  } & \text{gaussian}(y;0.8,-2)
\end{array}  
\right.
\end{equation}$$  

- 2.Output
$$\begin{equation}
Out = \{ (z,\mu_C(z))| z\in[-20,20]\} 
\end{equation}$$

$$\begin{equation}
\mu_C(z)=
\left\{
\begin{array}{}
\text{Positive:  } & \text{trapezoidal}(z;1,9,11,19) \\
\text{Zero:  } & \text{trapezoidal}(z;-9,-1,1,9) \\
\text{Negative:  } & \text{trapezoidal}(z;-19,-11,-9,-1)
\end{array}  
\right.
\end{equation}$$
