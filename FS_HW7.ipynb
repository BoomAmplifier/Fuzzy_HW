**Solution:**    
According to Figure E11.2-i, we have
$$\begin{equation}
(p,t)=\left[
\begin{array}{lll}
(-2,-2),&(-1.5,-2),&(-1,-1),&(-0.5,0), \\
(0,0),&(0.5,0),&(1,1),&(1.5,2),&(2,2)
\end{array}
\right]
\end{equation}$$

- STEP1: Chose initial values of the weights, biases, and input
$$\begin{equation}
W^1_{0}=
\begin{bmatrix}
-0.27 \\ -0.41
\end{bmatrix}
,b^1_{0}=
\begin{bmatrix}
-0.48 \\ -0.13
\end{bmatrix}
, W^2_{0}=
\begin{bmatrix}
-0.09 & -0.17
\end{bmatrix}
,b^2_{0}=
\begin{bmatrix}
0.48
\end{bmatrix}
\end{equation}$$
$$p^0=-1.5, t^0=-2$$
$$a^0=p^0$$

- STEP2: Calculate the outputs and error    

The output of the first layer     
$$\begin{equation}
a^1=f^1(W^1_0 a^0 + b^1_0)=
\rm{logsig}\left(
\begin{bmatrix}
-0.27 \\ -0.41
\end{bmatrix}
[-1.5]+
\begin{bmatrix}
-0.48 \\ -0.13
\end{bmatrix}
\right)=
\begin{bmatrix}
0.4812 \\ 0.6189
\end{bmatrix}
\end{equation}$$     
The output of the second layer     
$$\begin{equation}
a^2=f^2(W^2_0 a^1 + b^2_0)=
\rm{purelin}\left(
\begin{bmatrix}
0.09 & -0.17
\end{bmatrix}
\begin{bmatrix}
0.4812 \\ 0.6189
\end{bmatrix}
+[0.48]
\right)
=0.3315
\end{equation}$$     
The error is      
$$e=t-a=-2-0.3315=-2.3315$$

- STEP3: Calculate the derivatives of the transfer functions  
$$\dot{f}^1(n)=\frac{\rm d}{\rm d n}(\frac{1}{1+e^{-n}}=(1-a^1)(a^1)$$
$$\dot{f}^2(n)=\frac{\rm d}{\rm d n}(n)=1$$

- STEP4: Perform the backpropagation
$$\begin{equation}
\begin{array}{lll}
s^2&=-2\dot{F}^2(n^2)(t-a)=-2[1](-2.3315)=4.663\\
s^1 &= -2\dot{F}^1(n^1)(W^2)^Ts^2\\
&=
\begin{bmatrix}
(1-a^1_1)(a^1_1) & 0 \\
0 & (1-a^1_1)(a^1_1)
\end{bmatrix}
\begin{bmatrix}
-0.09 \\ -0.17
\end{bmatrix}
[4.663] \\
&=
\begin{bmatrix}
(1-0.321)(0.321) & 0 \\
0 & (1-0.368)(0.368)
\end{bmatrix}
\begin{bmatrix}
-0.09 \\ -0.17
\end{bmatrix}
[4.663]\\
&=
\begin{bmatrix}
-0.0915 \\ -0.1844
\end{bmatrix}
\end{array}
\end{equation}$$

- STEP5: Update the weights, let $\alpha=0.1$
$$\begin{equation}
\begin{array}{lll}
W^2_{1} &= W^2_{0}-\alpha s^2(a^1)^T \\
&= 
\begin{bmatrix}
-0.09 & -0.17
\end{bmatrix}
-0.1[4.663]
\begin{bmatrix}
-0.48 & -0.13
\end{bmatrix}\\
&=
\begin{bmatrix}
0.1338 & -0.1094
\end{bmatrix}\\
b^2_1&=b^2_0-\alpha s^2\\
&=[0.48]-0.1[4.663]=0.0137\\
W^1_{1} &= W^1_{0}-\alpha s^1(a^0)^T \\
&= 
\begin{bmatrix}
-0.27 \\ -0.41
\end{bmatrix}
-0.1
\begin{bmatrix}
-0.0915 \\ -0.1844
\end{bmatrix}
[-1.5]\\
&=
\begin{bmatrix}
-0.2837 \\ -0.4377
\end{bmatrix}\\
b^1_1&=b^1_0-\alpha s^1 \\
&=
\begin{bmatrix}
-0.48 \\ -0.13
\end{bmatrix}
-0.1
\begin{bmatrix}
-0.0915 \\ -0.1844
\end{bmatrix}\\
&=
\begin{bmatrix}
-0.4709 \\ -0.1116
\end{bmatrix}
\end{array}
\end{equation}$$
- STEP6: Chose another point, perform backpropagation again.
Finally, the weights and biases are as follows
$$\begin{equation}
W^1=
\begin{bmatrix}
10 \\ 10
\end{bmatrix}
,b^1=
\begin{bmatrix}
10 \\ -10
\end{bmatrix}
, W^2=
\begin{bmatrix}
2 & 2
\end{bmatrix}
,b^2=
\begin{bmatrix}
-2
\end{bmatrix}
\end{equation}$$

#### Problem 2 (Handout-E11.6):  
For the network shown in Figure E11.4 the initial weights and biases are chosen to be 
$$w^1(0)=1, b^1(0)=-2, w^2(0)=1, b^2(0)=1$$
The network transfer functions are 
$$f^1(n)=(n)^2, f^2(n)=\frac{1}{n}$$
and an input/target pair is given to be 
$${p=1,t=1}$$
Perform one iteration of backpropagation with $\alpha =1$.

<div>
<center>
    <img src=".\fig\hw7-2.jpg" style="width: 60%;height: 60%"/>
</center>
</div> 


**Solution:** 
- STEP1: Calculate the outputs and error
The initial value    
$$a^0 = p=1$$

The output of the first layer   

$$n^1=w^1_0 a^0 + b^1_0 =1 \times 1 + (-2)=-1$$

$$\begin{equation}
a^1=f^1(n^1)=
\left(
-1
\right)^2
= 1
\end{equation}$$

The output of the second layer

$$n^2=w^2_0 a^1 + b^2_0 =1 \times 1 +1=2$$

$$\begin{equation}
a^2=f^2(w^2_0 a^1 + b^2_0)=
\left(
2
\right)^{-1}
=0.5
\end{equation}$$

The error is 

$$e=t-a=1-0.5=0.5$$

- STEP2: Calculate the derivatives of the transfer functions       
$$\dot{f}^1(n)=\frac{\rm d}{\rm d n}(n^2)=2n$$
$$\dot{f}^2(n)=\frac{\rm d}{\rm d n}(\frac{1}{n})=-\frac{1}{n^2}$$

- STEP3: Perform the backpropagation
$$\begin{equation}
\begin{array}{lll}
s^2&=-2\dot{F}^2(n^2)(t-a)=-2[-\frac{1}{2^2}](0.5)=0.25\\
s^1 &= -2\dot{F}^1(n^1)(w^2)^Ts^2 = -2[2\times -1]\times 1 \times 0.25=1
\end{array}
\end{equation}$$

- STEP4: Update the weights, let $\alpha=1$
$$\begin{equation}
\begin{array}{lll}
w^2_{1} &= w^2_{0}-\alpha s^2(a^1)^T =1-1\times 0.25 \times 1=0.75 \\
b^2_1&=b^2_0-\alpha s^2 = 1-1\times 0.25 =0.75\\
w^1_{1} &= w^1_{0}-\alpha s^1(a^0)^T =1-1\times 1\times 1=0\\
b^1_1&=b^1_0-\alpha s^1 =-1\\
\end{array}
\end{equation}$$
