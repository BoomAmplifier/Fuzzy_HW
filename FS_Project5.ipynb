- Genetic algorithm      

Using the code shown as below to search minimum points of the functhion:      

>```
fun = @(x)p3f(x(1),x(2))
options = optimoptions('ga','ConstraintTolerance',1e-6,'PlotFcn', @gaplotbestf);
rng default % For reproducibility
x = ga(fun,2,A,b,Aeq,beq,lb,ub,[],options)
>```                
                
When the searching boundary $lb,ub$ is defined, the algorithm reachs the same global minimum/maximum points as same as Derivative-based Optimization.      
By changing different searching boundaries, local global minimum/maximum points could be found.
