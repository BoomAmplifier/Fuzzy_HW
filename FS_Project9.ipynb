## Project 9      
The Matlab code for Project8 is in the attached file `YQ_FS_P9.mlx`. The results are as follows in short.   

$$\begin{equation}
\begin{array}{rcl}
f(x)&=& \frac{\sum^{2}_{i=1} q^i \mu_i(x)}{\sum^{2}_{i=1}\mu_i(x)} \\
&=& [a_0^1+a_1^1 x]\xi_1(x) + [a_0^2+a_1^2 x]\xi_2(x) \\
&=& \phi^T\theta
\end{array}
\end{equation}$$


$$\begin{equation}
\phi^T = 
\begin{bmatrix}
\xi_1(x) & x\xi_1(x) & \xi_2(x) & x\xi_2(x) 
\end{bmatrix}
\end{equation}$$
$$\begin{equation}
\theta = 
\begin{bmatrix}
a_0^1 &  a_0^2 & a_1^1 & a_1^2
\end{bmatrix}
\end{equation}$$
