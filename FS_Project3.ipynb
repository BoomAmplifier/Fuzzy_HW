## Solutions:  

- Construct a polynomial $p(x)$ as the target function:     

$$ p_n(x) = p_0 + p_1x^1 + p_2x^2+ \cdots + p_nx^n$$

- Estimate each parameters for 1st-order, 2nd-order, 3rd-order, and 4th-order polynomials:     

$$\begin{equation}
\left\{
\begin{array}{lcl}
p_1(x) &=& 1.1986 + 0.444x \\
p_2(x) &=& 0.7588 + 0.6898x - 0.0242x^2 \\
p_3(x) &=& 1.1181 + 0.3520x + 0.0542x^2 -0.0051x^3 \\
p_4(x) &=& 1.0118 + 0.4908x + 0.0005x^2 +0.0028x^3-0.0004x^4
\end{array}
\right.
\end{equation}
$$

force = [1.1; 1.9; 3.2; 4.4; 5.9; 7.4; 9.2];
leng  = [1.5; 2.1; 2.5; 3.3; 4.1; 4.6; 5.0];
x0 = (0:0.5:11);
%
fun1 = @(x,force)( x(1)+x(2).*force  ); %x(1) +
fun2 = @(x,force)( x(1)+x(2).*force + x(3).*force.^2  );
fun3 = @(x,force)( x(1)+x(2).*force + x(3).*force.^2 + x(4).*force.^3 );
fun4 = @(x,force)( x(1)+x(2).*force + x(3).*force.^2 + x(4).*force.^3 + x(5).*force.^4 );
%
options = optimoptions('lsqcurvefit','Algorithm','levenberg-marquardt');
lb = [];
ub = [];
x_start=[0,0,0,0,0];
lsqf1 = lsqcurvefit(fun1, x_start(1:2), force, leng, lb, ub, options);
lsqf2 = lsqcurvefit(fun2, x_start(1:3), force, leng, lb, ub, options);
lsqf3 = lsqcurvefit(fun3, x_start(1:4), force, leng, lb, ub, options);
lsqf4 = lsqcurvefit(fun4, x_start, force, leng, lb, ub, options);
%
disp('1st-order:')
disp(lsqf1)
disp('2nd-order:')
disp(lsqf2)
disp('3rd-order:')
disp(lsqf3)
disp('4th-order:')
disp(lsqf4)
%
figure(1)
subplot(221);
plot(x0, fun1(lsqf1,x0), '-', force, leng, '*');
axis([0 10 1 5.5]);
xlabel('Force'); ylabel('Length of Spring');
title('(a) First-order Polynomial');
subplot(222);
plot(x0, fun2(lsqf2,x0), '-', force, leng, '*');
axis([0 10 1 5.5]);
xlabel('Force'); ylabel('Length of Spring');
title('(b) Second-order Polynomial');
%
subplot(223);
plot(x0, fun3(lsqf3,x0), '-', force, leng, '*');
axis([0 10 1 5.5]);
xlabel('Force'); ylabel('Length of Spring');
title('(c) Third-order Polynomial');
%
subplot(224);
plot(x0, fun4(lsqf4,x0), '-', force, leng, '*');
axis([0 10 1 5.5]);
xlabel('Force'); ylabel('Length of Spring');
title('(d) Fourth-order Polynomial');
```
