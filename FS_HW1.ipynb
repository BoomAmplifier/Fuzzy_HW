**Solution:**    
- 1.$A\cap \bar{A}=\emptyset$       

$\forall A$, $A$ is a fuzzy set, its complement fuzzy set is $\bar{A}$. According to the definition of fuzzy set, $\mu_A(x)$ describes the degree to which $x$ belongs to A while $\mu_{\bar{A}}(x)$ illustrates the degree to which $x$ not belongs to $A$. Due to $A\cap \bar{A}$ is defined as the degree to which $x$ belongs either $A$ or $\bar{A}$, there definitely no any $x$ could either belongs or not belongs a same $A$. Thus         
$$ A\cap \bar{A}= \text{min}\left(\mu_A(x),\mu_{\bar{A}}(x)\right)= \text{min}\left(\mu_A(x),1-\mu_A(x)\right)=\emptyset$$  
- 2.$ A \cup \bar{A} = X $      

The same explanation as $A\cap \bar{A}=\emptyset$, $\forall x$ will belong or not belong to a $A$. That means    
$$ A\cup \bar{A}= \text{max}\left(\mu_A(x),\mu_{\bar{A}}(x)\right)= \text{max}\left(\mu_A(x),1-\mu_A(x)\right)=X$$  
- 3.$ A \cap A =A , A \cup A = A $     

$$ A \cap A = \text{min}\left(\mu_A(x),\mu_A(x)\right) = \mu_A(x) = A $$
$$ A \cup A = \text{max}\left(\mu_A(x),\mu_A(x)\right) = \mu_A(x) = A $$  
- 4.$ \bar{\bar{A}} = A $     

$$ \bar{\bar{A}} = 1-(1-\mu_A(x)) = \mu_A(x) =A $$  
- 5.$ A\cap B= B \cap A, A\cup B= B \cup A $     

$$ A\cap B= \text{min}\left(\mu_A(x),\mu_B(x)\right) = \text{min}\left(\mu_B(x),\mu_A(x)\right)$$
$$ A\cup B= \text{max}\left(\mu_A(x),\mu_B(x)\right) = \text{max}\left(\mu_B(x),\mu_A(x)\right)$$
- 6.$ (A\cup B)\cup C = A \cup (B\cup C)$ , $ (A\cap B)\cap C = A \cap (B\cap C)  $

$$ \begin{equation}
\begin{array}{rCl}
(A\cup B)\cup C &= \text{max}\left( \text{max}(\mu_A(x),\mu_B(x)),\mu_C(x)\right) \\
&= \text{max}\left( \mu_A(x),\mu_B(x),\mu_C(x)\right) \\
&= \text{max}\left( \mu_A(x),\text{max}(\mu_B(x),\mu_C(x))\right) \\
&= A\cup (B\cup C)
\end{array}
\end{equation}
$$
$$ \begin{equation}
\begin{array}{rCl}
(A\cap B)\cup C &= \text{min}\left( \text{min}(\mu_A(x),\mu_B(x)),\mu_C(x)\right) \\
&= \text{min}\left( \mu_A(x),\mu_B(x),\mu_C(x)\right) \\
&= \text{min}\left( \mu_A(x),\text{min}(\mu_B(x),\mu_C(x))\right) \\
&= A\cap (B\cap C)
\end{array}
\end{equation}
$$
- 7.$ A\cup (B\cap C )= (A\cup B)\cap(A\cup C)$ ， $ A\cap (B\cup C )= (A\cap B)\cup(A\cap C)  $       

$$ \begin{equation}
\begin{array}{rCl}
A\cup (B\cap C) &= \text{max}\left( \mu_A(x),\text{min}(\mu_B(x),\mu_C(x))\right) \\
&= \left\{
\begin{array}{rCl}
\text{max}\left( \mu_A(x),\mu_B(x)\right) &\text{  if  } \mu_B(x)<\mu_C(x)\\ 
\text{max}\left( \mu_A(x),\mu_C(x)\right) &\text{  if  } \mu_B(x)>\mu_C(x)
\end{array}
\right.
\\  
&= \text{max}\left(\text{max}( \mu_A(x),\mu_B(x)),\text{max}( \mu_A(x),\mu_C(x))\right) \\
&= (A\cup B)\cap(A\cup C)
\end{array}
\end{equation}
$$
$$ \begin{equation}
\begin{array}{rCl}
A\cap (B\cup C ) &= \text{min}\left( \mu_A(x),\text{max}(\mu_B(x),\mu_C(x))\right) \\
&= \left\{
\begin{array}{rCl}
\text{min}\left( \mu_A(x),\mu_B(x)\right) &\text{  if  } \mu_B(x)>\mu_C(x)\\   
\text{min}\left( \mu_A(x),\mu_C(x)\right) &\text{  if  } \mu_B(x)<\mu_C(x)
\end{array}
\right.
\\
&= \text{min}\left(\text{min}( \mu_A(x),\mu_B(x)),\text{min}( \mu_A(x),\mu_C(x))\right) \\
&= (A\cap B)\cup(A\cap C)
\end{array}
\end{equation}
$$
- 8.$ A\cup (A\cap B )= A$ ， $ A\cap (A\cup B )= A$       

$$ \begin{equation}
\begin{array}{rCl}
A\cup (A \cap B ) &= \text{max} \left( \text{min}( \mu_A(x),\mu_B(x)\right) \\
&= \left\{
\begin{array}{rCl}
\text{max}\left( \mu_A(x),\mu_A(x)\right)= \mu_A(x) &\text{  if  } \mu_A(x)<\mu_B(x)\\   
\text{max}\left( \mu_A(x),\mu_B(x)\right)= \mu_A(x) &\text{  if  } \mu_A(x)>\mu_B(x)
\end{array}
\right.
\\
&= A
\end{array}
\end{equation}
$$
$$ \begin{equation}
\begin{array}{rCl}
A\cap (A\cup B) &= \text{min} \left( \text{max}( \mu_A(x),\mu_B(x)\right) \\
&= \left\{
\begin{array}{rCl}
\text{min}\left( \mu_A(x),\mu_A(x)\right)= \mu_A(x) &\text{  if  } \mu_A(x)>\mu_B(x)\\   
\text{min}\left( \mu_A(x),\mu_B(x)\right)= \mu_A(x) &\text{  if  } \mu_A(x)<\mu_B(x)
\end{array}
\right.
\\
&= A
\end{array}
\end{equation}
$$
- 9.$ A\cup (\bar{A}\cap B )= A \cap B $ , $ A\cap (\bar{A}\cup B )= A \cup B $    

$$ \begin{equation}
\begin{array}{lCl}
A\cup (\bar{A}\cap B ) &= \text{max} \left( \text{min}(1-\mu_A(x),\mu_B(x)\right) \\
&= \left\{
\begin{array}{lCl}
\text{max}\left( \mu_A(x),\mu_B(x)\right) &\text{  if  } 1-\mu_A(x)>\mu_B(x)\\   
\text{max}\left( \mu_A(x),1-\mu_A(x)\right)= X   &\text{  if  } 1-\mu_A(x)<\mu_B(x)
\end{array}
\right.
\\
&= A \cap B 
\end{array}
\end{equation}
$$
$$ \begin{equation}
\begin{array}{rCl}
A\cap (\bar{A}\cup B ) &= \text{min} \left( \text{max}(1-\mu_A(x),\mu_B(x)\right) \\
&= \left\{
\begin{array}{}
\text{min}\left( \mu_A(x),\mu_B(x)\right) &\text{  if  } 1-\mu_A(x)<\mu_B(x)\\   
\text{min}\left( \mu_A(x),1-\mu_A(x)\right)= \emptyset   &\text{  if  } 1-\mu_A(x)>\mu_B(x)
\end{array}
\right.
\\
&= A \cup B 
\end{array}
\end{equation}
$$
- 10.$\overline{A\cup B}=\bar{A}\cap\bar{B}$ , $\overline{A\cap B}=\bar{A}\cup\bar{B}$   

$$ \begin{equation}
\begin{array}{lCl}
\forall x \in \overline{A\cup B} &\Rightarrow x \notin A\cup B \\
&\Rightarrow x \notin A \text{ or } x \notin B \\
&\Rightarrow x \in \bar{A} \text{ or } x \in \bar{B} \\
&\Rightarrow x \in \bar{A}\cap \bar{B}
\end{array}
\end{equation}$$    

$$\begin{equation}
\begin{array}{rCl}
\forall x \in \bar{A}\cup\bar{B} &\Rightarrow x \in \bar{A} \cup \bar{B} \\
&\Rightarrow x \in \bar{A} \text{ or } x \in \bar{B} \\
&\Rightarrow x \notin A \text{ or } x \notin B \\
&\Rightarrow x \notin A \cap B \\
&\Rightarrow x \in \overline{A\cap B}
\end{array}
\end{equation}$$

#### Problem 2 (Ch2-5):    
Suppose that fuzzy set $A$ is described by $\mu_A(x)=\text{bell}(x; a, b, c)$. Show that the classical fuzzy complement of $A$ is described by $\mu_{\bar{A}}(x)=\text{bell}(x; a, -b, c)$.

**Solution:**     
According to (2.23)    
$$\mu_A(x)=\frac{1}{1+\left| \frac{x-c}{a} \right|^{2b} }$$
$$\begin{equation}
\begin{array}{rCl}
\mu_\bar{A}(x) &= 1-\frac{1}{1+\left| \frac{x-c}{a} \right|^{2b} }   \\
&= \frac{\left| \frac{x-c}{a} \right|^{2b}}{1+\left| \frac{x-c}{a} \right|^{2b} } \\
&= \frac{1}{\frac{1}{\left| \frac{x-c}{a} \right|^{2b}}+1 } \\
&= \frac{1}{1+\left| \frac{x-c}{a} \right|^{-2b} } \\
&= \text{bell}(x; a, -b, c)
\end{array}
\end{equation}$$
Draw $\text{bell}(x; 5, 2, 1)$ and $\text{bell}(x; 5, -2, 1)$ in python3, the figure shows the relationship clearly.

```python
import numpy as np
import matplotlib.pyplot as plt
import random
import skfuzzy as fuzz
%matplotlib inline

x = np.arange(11)
mfx= fuzz.gbellmf(x,5.0,2,1)
n_mfx= fuzz.gbellmf(x,5.0,-2,1)
plt.plot(x,mfx)
plt.plot(x,n_mfx)
plt.show()
```

#### Problem 4 (Ch2-11):    
Find the level set $A_{\alpha}$ and its width for a fuzzy set $A$ defined by $\mu_A(x)=\text{trapezoid}(x; a, b, c, d).$   

**Solution:**    
$$\begin{equation}
\mu_A(x) =  
\left\{
\begin{array}{}
0,  & x \leq a \\
\frac{x-a}{b-a}, & a\leq x \leq b \\
1, & b \leq x \leq c \\
\frac{d-x}{d-c}, & c \leq x \leq d \\
0, & b\leq d \leq x
\end{array}
\right.
\end{equation}
$$
The level set $A_{\alpha}$ is defined as   
$$\begin{equation}
\left\{
\begin{array}{}
\frac{x_1-a}{b-a} = \alpha \\
\frac{d-x_2}{d-c} = \alpha
\end{array}
\right.
\Rightarrow
\left\{
\begin{array}{}
x_1 = a+(b-a)\alpha \\
x_2 = d+(c-d)\alpha
\end{array}
\right.
\end{equation}
$$
The crossover points are: $(a+(b-a)\alpha, \alpha),(d+(c-d)\alpha, \alpha)$.     
The width of $A$ and $A_{\alpha}$  is:   
$$\text{Width}_{A_{\alpha}-A} = x_2-x_1 = d-a+(c-d-b+a)\alpha$$


```python
x = np.arange(100)
a = random.randint(0, 15)
b = random.randint(15, 50)
c = random.randint(50, 85)
d = random.randint(85, 100)

alph1 = 0.5
alph2 = 0.8

A_alph1 = alph1 + 0*x
A_alph2 = alph2 + 0*x

mfx= fuzz.trapmf(x,[a,b,c,d])
plt.plot(x,mfx)
plt.plot(x,A_alph1)
plt.plot(x,A_alph2)
plt.show()
```

#### Problem 5 (Ch2-12):    
Derive the partial derivatives of a Gaussian MF $y=\text{gaussian}(x; c, \sigma)$ with respect to its argument $x$ and parameters $\sigma$ and $c$, and thus verify Equations (2.29) to (2.31).

**Solution:**   
$$y=e^{-\frac{1}{2}(\frac{x-c}{\sigma})^2}$$
According to 
$$\frac{\partial{y}}{\partial{(x/c/\sigma})} = e^{-\frac{1}{2}(\frac{x-c}{\sigma})^2} \cdot \frac{\partial{\frac{x-c}{\sigma})^2}}{\partial{(x/c/\sigma})} = \frac{\partial{\frac{x-c}{\sigma})^2}}{\partial{(x/c/\sigma)}}y$$
Thus 
1. $\frac{\partial{y}}{\partial{x}} = \frac{\partial\left[{-\frac{1}{2}(\frac{x-c}{\sigma})^2}\right]}{\partial{x}} = -\frac{x-c_1}{\sigma_1}\frac{\partial{\frac{x-c}{\sigma_1}}}{\partial{x}} = -\frac{x-c}{\sigma ^2}y$
2. $\frac{\partial{y}}{\partial{\sigma}} = \frac{\partial\left[ -\frac{1}{2} (x-c)^2 \sigma^{-2} \right]}{\partial{\sigma}} = \frac{(x-c)^2}{\sigma^{3}}y$
3. $\frac{\partial{y}}{\partial{c}} = \frac{\partial\left[{-\frac{1}{2}(\frac{x-c}{\sigma})^2}\right]}{\partial{c}} = \frac{x-c}{\sigma ^2}y$

#### Problem 6 (Ch2-14):    
Let the fuzzy set $A$ be defined by a Gaussian MF $\text{gaussian}(x, c, \sigma)$. Show that $\text{width}(A_{0.99})/\text{width}(A_{0.01})$ is a constant independent of the parameters $c$ and $\sigma$.

**Solution:**
$$A=e^{-\frac{1}{2}(\frac{x-c}{\sigma})^2}$$
$$\begin{equation}
\begin{array}{rCl}
&A_{\alpha} &= \left\{ x; e^{-\frac{1}{2}(\frac{x-c}{\sigma})^2} \geq \alpha \right\} \\
&\Rightarrow e^{-\frac{1}{2}(\frac{x-c}{\sigma})^2} &= \alpha \\
&\Rightarrow (\frac{x-c}{\sigma})^2 &= -2\ln{\alpha} \\
&\Rightarrow x &= \pm \sigma \sqrt{-2\ln{\alpha}} + a
\end{array}
\end{equation}$$
Thus   
$$ \text{Width}_{A_{\alpha}-A} = \sigma \sqrt{-2\ln{\alpha}} + a +\sigma \sqrt{-2\ln{\alpha}} -a = 2\sigma \sqrt{-2\ln{\alpha}} $$
Consquently
$$ \frac{\text{Width}_{A_{0.99}-A}}{\text{Width}_{A_{0.01}-A}}= \frac{2\sigma \sqrt{-2\ln{0.99}}}{2\sigma \sqrt{-2\ln{0.01}}}=\sqrt{\frac{ln{0.99}}{ln{0.01}}}$$

#### Problem 7 (Ch2-17):    
Show that the Sugeno's and Yager's complement operators (Examples 2.10 and 2.11) satisfy the involution requirement of Equation (2.38).

**Solution:**  
- Sugeno's complement
$$\begin{equation}
\begin{array}{rCl}
N_s(N_s(a)) &= \frac{1-\frac{1-a}{1+sa}}{1+s\frac{1-a}{1+sa}} \\
&= \frac{1+sa-1+a}{1+sa+s-sa} \\
&= \frac{(1+s)a}{1+s} \\
&= a
\end{array}
\end{equation}
$$
- Yager's complement
$$\begin{equation}
\begin{array}{rCl}
N_w(N_w(a)) &= \left[ 1- \left[ (1-a^w)^{1/w} \right]^w \right]^{1/w} \\
&= \left[ 1- (1-a^w) \right]^{1/w} \\
&= a
\end{array}
\end{equation}
$$

#### Problem 8 (Ch2-18):    
Verify that the four T-norm and T-conorm operators in Examples 2.12 and 2.13 are dual to each other in the sense of the generalized DeMorgan's law.

**Solution:**  
Choose $N(z)=1-z$ as the fuzzy complement operator.
- 1.$T_{min}$ and $S_{max} $   
$$\begin{equation}
\begin{array}{rCl}
T_{min}(a,b) &= N(N(T_{min}(a, b)))  \\
&= 1-(1-T_{min}(a, b)) \\
&= 1-(1-\text{min}(a, b))  \\
&= 1- \text{max}(1-a, 1-b) \\
&= 1- S_{max}(1-a, 1-b)
\end{array}
\end{equation}
$$

$$\begin{equation}
\begin{array}{rCl}
S_{max}(a,b) &= N(N(S_{max}(a, b)))  \\
&= 1-(1-T_{max}(a, b)) \\
&= 1-(1-\text{max}(a, b))  \\
&= 1- \text{min}(1-a, 1-b) \\
&= 1- T_{min}(1-a, 1-b)
\end{array}
\end{equation}
$$
- 2.$T_{ap}$ and $S_{ap} $ 
$$\begin{equation}
\begin{array}{rCl}
T_{ap}(a,b) &= N(N(T_{ap}(a, b)))  \\
&= 1-(1-T_{ap}(a, b)) \\
&= 1-(1-ab)  \\
&= 1- (1-a+ 1-b -(1-a)(1-b)) \\
&= 1- S_{ap}(1-a, 1-b)
\end{array}
\end{equation}
$$

$$\begin{equation}
\begin{array}{rCl}
S_{ap}(a,b) &= N(N(S_{ap}(a, b)))  \\
&= 1-(1-S_{ap}(a, b)) \\
&= 1-(1-a-b+ab)  \\
&= 1- (1-a)(1-b) \\
&= 1- T_{ap}(1-a, 1-b)
\end{array}
\end{equation}
$$
- 3.$T_{bp}$ and $S_{bp} $ 
$$\begin{equation}
\begin{array}{rCl}
T_{bp}(a,b) &= N(N(T_{bp}(a, b)))  \\
&= 1-(1-T_{bp}(a, b)) \\
&= 1-(1-0\cup (a+b-1))  \\
&= 1- (1\cap (1-a+1-b)) \\
&= 1- S_{bp}(1-a, 1-b)
\end{array}
\end{equation}
$$

$$\begin{equation}
\begin{array}{rCl}
S_{bp}(a,b) &= N(N(S_{bp}(a, b)))  \\
&= 1-(1-S_{np}(a, b)) \\
&= 1-(1-1\cap (a+b))  \\
&= 1- (0\cup (1-a-b)) \\
&= 1- (0\cup (1-a+1-b-1)) \\
&= 1- T_{bp}(1-a, 1-b)
\end{array}
\end{equation}
$$
- 4.$T_{dp}$ and $S_{dp} $ 
$$\begin{equation}
\begin{array}{rCl}
T_{dp}(a,b) &= N(N(T_{dp}(a, b))) \\
&= \left\{ 
\begin{array}{rCl}
1-(1-a) &  b=1 &\Rightarrow 1-b=0 \\
1-(1-b) &  a=1 &\Rightarrow 1-a=0 \\
1-(1-0) &  a,b<1 &\Rightarrow (1-a),(1-b)>0 
\end{array}
\right. \\
&= 1-S_{dp}(1-a,1-b)
\end{array}
\end{equation}
$$

$$\begin{equation}
\begin{array}{rCl}
S_{dp}(a,b) &= N(N(S_{dp}(a, b))) \\
&= \left\{ 
\begin{array}{rCl}
1-(1-a) &  b=0 &\Rightarrow 1-b=1 \\
1-(1-b) &  a=0 &\Rightarrow 1-a=1 \\
1-(1-0) &  a,b>1 &\Rightarrow (1-a),(1-b)<1
\end{array}
\right. \\
&= 1-T_{dp}(1-a,1-b)
\end{array}
\end{equation}
$$
